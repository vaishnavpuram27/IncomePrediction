```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Per_Capita_Income.csv")
df.head
```

```python
df.columns
```

```python
df.iloc(2)
```

```python
df["year"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df[["year"]]
```

```python
y = df[["income"]]
```

```python
x 
```

```python
y
```

```python
reg = LinearRegression().fit(x, y)
```

```python
reg.score(x,y)
```

```python
reg.predict([[2022]])[0][0]
```

```python
import pickle
```

```python
filename = "model.pkl"
pickle.dump(reg, open(filename, 'wb'))
```

```python
loadModel = pickle.load(open(filename, 'rb'))
```

```python
result = loadModel.predict([[2022]])[0][0]
print(result)
```
